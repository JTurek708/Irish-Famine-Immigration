### Famine Irish Ships Text file notes
- The README specifies the Irish ships txt file as FIMHR
    - FIMHR = Famine Irish Manifest Header File
    - 3,226 records - one record per manifest

- Text file has 5 columns:
    - Col 1: CIR-MID: Key to records in FIPAS file
        - FIPAS = Famine Irish Passenger Record File
            - 604,596 records - 1 per passenger
    - Col 2: Ship Name
    - Col 3: CIR - Code for Ship Port of Embarkation
        - See FIPRT file for code (Famine Irish Port Code)
    - Col 4: Ship Arrival Date
    - Col 5: Number of Corresponding Passenger Records in FIPAS file

### Famine Irish Passenger Record File (FIPAS)
- Refer to README for different codes 
    -   Family Relation Code
    - Age 
    - CIR = Occupation Code
        - See FIOCC file for code
        - This is a PDF which will require parsing...
    - CIR = Native Country Code
        - SEE FICTY file code
    - CIR - Passenger Port of Embarkation Code
